```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
from sklearn.metrics.pairwise import cosine_similarity
from ast import literal_eval
from random import sample
```

```python
#Read the files in our device by providing the path of the file to read_csv function.
credits_df=pd.read_csv(r"C:\Users\Saisree\Downloads\archive\tmdb_5000_credits.csv")
movies_df=pd.read_csv(r"C:\Users\Saisree\Downloads\archive\tmdb_5000_movies.csv")
```

```python
#Merge the two dataframes 
credits_df.columns=['id','title','cast','crew']
movies_df=movies_df.merge(credits_df,on='id')
```

```python
#---------------------Recommending movies based on director,cast,crew,keywords of the inputed movie--------------------------
```

```python
features=["cast","crew","keywords","genres"]

for feature in features:
    movies_df[feature]=movies_df[feature].apply(literal_eval)
#applying literal_eval on each and every feature.
#Before applying literal_eval type of crew is List
#after applying literal_eval type of crew is dictionary.
movies_df[features].head(10)
```

```python
#Defining get_director function
def get_director(x):
    for i in x:
        if i["job"]=='Director':
            return i["name"]
    return np.nan
```

```python
#defining get_list function
def get_list(x):
    if isinstance(x,list):
        names=[i["name"] for i in x]
        if len(names)>3:
            names=names[:3]
        return names
    return []
```

```python
#getting or extracting the director name from the crew
movies_df["director"]=movies_df["crew"].apply(get_director)

#getting the first three elements for each feature.Example:getting three actor names in cast,three genres to which movie belong...
features = ["cast", "keywords", "genres"]
for feature in features:
    movies_df[feature]=movies_df[feature].apply(get_list)
    
```

```python
#movies_df[['original_title', 'cast', 'director', 'keywords', 'genres']].head()
features=['original_title', 'cast', 'director', 'keywords', 'genres']
movies_df[features].head()
```

```python
#cleaning the data =>removing white spaces
def clean_data(x):
    if isinstance(x,list):
        return [str.lower(i.replace(" ","")) for i in x]
    else:
        if isinstance(x,str):
            return str.lower(x.replace(" ",""))
        else:
            return " "
```

```python
features=["cast","keywords","director","genres"]
for feature in features:
    movies_df[feature]=movies_df[feature].apply(clean_data)
```

```python
features=['original_title', 'cast', 'director', 'keywords', 'genres']
movies_df[features].head()
```

```python
#create a “soup” containing all of the metadata information extracted to input into the count_vectorizer
def create_soup(x):
     return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])
    

movies_df["soup"]=movies_df.apply(create_soup,axis=1) #axis=1 => along the row
print(movies_df["soup"].head())
    
```

```python
count_vectorizer=CountVectorizer(stop_words="english")
#countVectorizer counts the frequency of each word and outputs a vector containing frequencies
count_matrix=count_vectorizer.fit_transform(movies_df["soup"])
print(count_matrix.shape)
#print the shape of the count_matrix formed after vectorizing the obtained soup using countvectorizer
cosine_sim=cosine_similarity(count_matrix,count_matrix)#cosine similarity measures the theta between the vectors inputed to it.
print(cosine_sim.shape)
#print the cosine similarity shape after finding the cosine similarity between count_matrix and count_matrix
movies_df=movies_df.reset_index()
indices=pd.Series(movies_df.index,index=movies_df["original_title"])
```

```python
#create a function that takes in the movie title and the cosine similarity score as input and outputs the movies similar to it.
def get_recommendations(title,cosine_sim=cosine_sim):
    idx=indices[title]
    sim_scores=list(enumerate(cosine_sim[idx]))
    sim_scores=sorted(sim_scores,key=lambda x:x[1],reverse=True)#arrange the similarity scores in descending order
    sim_scores=sim_scores[1:11] #take first 10 movies except the first one as it is the same movie that we inputed.
    movies_indices=[ind[0] for ind in sim_scores]
    movies=movies_df["original_title"].iloc[movies_indices]
    return movies
```

```python
print("Recommendations for Avengers")
print(get_recommendations("The Avengers", cosine_sim))
```
